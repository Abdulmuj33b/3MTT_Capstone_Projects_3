```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split # Not used in the final pipeline but good for initial splitting if needed
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Mount Google Drive to access the datasets
from google.colab import drive
drive.mount('/content/drive')

# Define the file paths for the training and testing datasets in Google Drive
train_path = '/content/drive/MyDrive/NeuroWell/train.csv'
test_path = '/content/drive/MyDrive/NeuroWell/test.csv'

# Load the datasets into pandas DataFrames
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# Store original test set Employee IDs for the final submission file
test_employee_ids = df_test['Employee ID']

# Combine train and test DataFrames for consistent preprocessing
# We add a placeholder 'Burn Rate' column to the test set before combining,
# as the test set does not have the target variable.
df_test['Burn Rate'] = np.nan
df_combined = pd.concat([df_train, df_test], ignore_index=True)

# --- Phase 2: Exploratory Data Analysis (EDA) ---

print("--- Phase 2: EDA ---")

# Display information about each dataset (data types, non-null counts)
print("\nTrain Dataset Info:")
df_train.info()
print("\nTest Dataset Info:")
df_test.info()
print("\nCombined Dataset Info:")
df_combined.info()

# Display descriptive statistics for numerical columns
print("\nTrain Dataset Description:")
print(df_train.describe())
print("\nTest Dataset Description:")
print(df_test.describe())

# Check for missing values in the combined dataset
print("\nMissing values in combined dataset:")
print(df_combined.isnull().sum())

# Visualize the distribution of numerical features using histograms
numerical_features = ['Designation', 'Resource Allocation', 'Mental Fatigue Score', 'Burn Rate']
df_combined[numerical_features].hist(bins=30, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Features") # Add a main title to the figure
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()

# Visualize the relationships between key numerical features and the target variable (Burn Rate)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1) # Create a subplot for Resource Allocation vs Burn Rate
sns.scatterplot(x='Resource Allocation', y='Burn Rate', data=df_train)
plt.title('Burn Rate vs Resource Allocation')

plt.subplot(1, 3, 2) # Create a subplot for Mental Fatigue Score vs Burn Rate
sns.scatterplot(x='Mental Fatigue Score', y='Burn Rate', data=df_train)
plt.title('Burn Rate vs Mental Fatigue Score')

plt.subplot(1, 3, 3) # Create a subplot for Designation vs Burn Rate
sns.boxplot(x='Designation', y='Burn Rate', data=df_train)
plt.title('Burn Rate vs Designation')

plt.tight_layout() # Adjust layout
plt.show()

# Visualize the distribution of categorical features using countplots
categorical_features = ['Gender', 'Company Type', 'WFH Setup Available']
for col in categorical_features:
  plt.figure(figsize=(8, 5))
  sns.countplot(x=col, data=df_combined)
  plt.title(f'Count of {col}')
  plt.show()

# Visualize the relationship between categorical features and the target variable (Burn Rate) using boxplots
for col in categorical_features:
  plt.figure(figsize=(8, 5))
  sns.boxplot(x=col, y='Burn Rate', data=df_train)
  plt.title(f'Burn Rate vs {col}')
  plt.show()

# --- Phase 3: Data Preprocessing ---

print("\n--- Phase 3: Data Preprocessing ---")

# Feature Engineering: Convert 'Date of Joining' to datetime and calculate 'Tenure_Days'
df_combined['Date of Joining'] = pd.to_datetime(df_combined['Date of Joining'], errors='coerce')
# Calculate tenure in days from the latest date in the combined dataset
latest_date_combined = df_combined['Date of Joining'].max()
df_combined['Tenure_Days'] = (latest_date_combined - df_combined['Date of Joining']).dt.days

# Drop the original 'Date of Joining' and 'Employee ID' columns as they are no longer needed after feature engineering
df_combined = df_combined.drop(['Date of Joining', 'Employee ID'], axis=1)

# Separate the combined data back into training and testing sets
# Use the original training indices to get the training data (dropping rows with missing Burn Rate)
df_train_processed = df_combined.iloc[train_indices].dropna(subset=['Burn Rate']).copy()
# Use the original test indices to get the test data
df_test_processed = df_combined.iloc[test_indices + len(df_train)].drop('Burn Rate', axis=1).copy()

# Define the features (X) and the target variable (y) for the training set
X_train = df_train_processed.drop('Burn Rate', axis=1)
y_train = df_train_processed['Burn Rate']
# The features for the test set are all columns except 'Burn Rate'
X_test = df_test_processed.copy()

# Identify numerical and categorical columns for preprocessing pipelines
numerical_cols = ['Designation', 'Resource Allocation', 'Mental Fatigue Score', 'Tenure_Days']
categorical_cols = ['Gender', 'Company Type', 'WFH Setup Available']

# Create a preprocessing pipeline for numerical features:
# 1. Impute missing values with the mean.
# 2. Scale features using StandardScaler.
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Create a preprocessing pipeline for categorical features:
# 1. Impute missing values with the most frequent value.
# 2. Apply One-Hot Encoding. Handle unknown categories by ignoring them during transformation.
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine the preprocessing steps using ColumnTransformer
# 'remainder='passthrough'' keeps any columns not specified in numerical_cols or categorical_cols (though in this case, all columns are covered)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough'
)

# --- Phase 4: Model Development ---

print("\n--- Phase 4: Model Development ---")

# Define the machine learning models to be used
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1), # n_jobs=-1 uses all available cores
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
}

results = {} # Dictionary to store evaluation results for each model
predictions = {} # Dictionary to store test predictions for each model

# Train and evaluate each model
for name, model in models.items():
    print(f"\nTraining {name}...")
    # Create a pipeline that first preprocesses the data and then applies the regressor model
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('regressor', model)])

    # Train the model pipeline on the training data
    pipeline.fit(X_train, y_train)

    # Make predictions on the training set for evaluation purposes
    y_train_pred = pipeline.predict(X_train)

    # Evaluate the model's performance on the training set using various metrics
    rmse = np.sqrt(mean_squared_error(y_train, y_train_pred)) # Root Mean Squared Error
    mae = mean_absolute_error(y_train, y_train_pred)       # Mean Absolute Error
    r2 = r2_score(y_train, y_train_pred)                  # R-squared

    # Store the evaluation metrics
    results[name] = {'RMSE': rmse, 'MAE': mae, 'R2': r2}

    # Print the evaluation metrics for the current model
    print(f"{name} - Training Metrics:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE: {mae:.4f}")
    print(f"  R2: {r2:.4f}")

    # Make predictions on the preprocessed test set
    test_predictions = pipeline.predict(X_test)
    # Clip the predictions to ensure they are within the valid range [0, 1] for burn rate
    test_predictions = np.clip(test_predictions, 0, 1)
    # Store the test predictions
    predictions[name] = test_predictions

# Print the evaluation results for all models
print("\nModel Evaluation Results (on Training Data):")
for name, metrics in results.items():
    print(f"{name}: RMSE={metrics['RMSE']:.4f}, MAE={metrics['MAE']:.4f}, R2={metrics['R2']:.4f}")

# Select the best performing model based on the R-squared score (higher is better)
best_model_name = max(results, key=lambda k: results[k]['R2'])
print(f"\nBest performing model based on R2: {best_model_name}")

# Get the test predictions from the best model
best_model_predictions = predictions[best_model_name]

# --- Phase 5: Insights, Recommendations and Solution Deployment ---

print("\n--- Phase 5: Insights and Recommendations ---")

# Analyze feature importance for tree-based models (Random Forest and Gradient Boosting)
if best_model_name in ['Random Forest', 'Gradient Boosting']:
    # Retrain the best model pipeline to ensure we have the fitted model for feature importance
    best_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                     ('regressor', models[best_model_name])])
    best_pipeline.fit(X_train, y_train)

    # Get the names of the features after preprocessing (including one-hot encoded features)
    # This requires accessing the fitted preprocessor within the pipeline
    ohe_feature_names = list(best_pipeline.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_cols))
    all_feature_names = numerical_cols + ohe_feature_names

    # Get the feature importances from the trained regressor model
    if hasattr(best_pipeline.named_steps['regressor'], 'feature_importances_'):
      importances = best_pipeline.named_steps['regressor'].feature_importances_
      # Create a DataFrame to display feature names and their importance scores
      feature_importance_df = pd.DataFrame({'feature': all_feature_names, 'importance': importances})
      # Sort the features by importance in descending order
      feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

      # Print the top 10 most important features
      print("\nFeature Importances (from best model):")
      print(feature_importance_df.head(10))

      # Visualize the top 10 feature importances
      plt.figure(figsize=(10, 6))
      sns.barplot(x='importance', y='feature', data=feature_importance_df.head(10))
      plt.title('Top 10 Feature Importances')
      plt.show()

# Create a DataFrame for the submission file with Employee IDs and the predicted Burn Rates from the best model
submission_df = pd.DataFrame({'Employee ID': test_employee_ids, f'{best_model_name}_Burn_Rate_Prediction': best_model_predictions})

# Save the submission DataFrame to a CSV file in Google Drive
submission_df.to_csv('/content/drive/MyDrive/NeuroWell/test_predictions.csv', index=False)

print(f"\nTest predictions from the best model ({best_model_name}) saved to /content/drive/MyDrive/NeuroWell/test_predictions.csv")
print("Sample of test predictions:")
print(submission_df.head())

# Insights and Recommendations based on the feature importance analysis and general knowledge about burnout
print("\nInsights and Recommendations:")
print("- Mental Fatigue Score and Resource Allocation are likely key drivers of burnout.")
print("  - Recommendation: Monitor employee stress levels and ensure balanced workload allocation.")
print("- Designation (seniority) and Tenure might also play a role, potentially indicating different pressures at different career stages.")
print("  - Recommendation: Tailor well-being support based on seniority and experience.")
print("- WFH Setup Availability could be significant. Lack of proper WFH setup may contribute to burnout.")
print("  - Recommendation: Provide resources and support for employees to set up comfortable and productive home workspaces.")
print("- Gender and Company Type might show differences in burnout rates.")
print("  - Recommendation: Investigate potential systemic issues or differences in work culture between company types or for different genders.")

print("\nFurther Recommendations:")
print("- Implement regular check-ins with employees, especially those with high predicted burnout risk.")
print("- Offer mental health resources, counseling, and stress management programs.")
print("- Encourage breaks and work-life balance.")
print("- Use predicted burnout scores to proactively identify employees who might need support.")

# Reflection Questions (as comments) - Addressing the reflection questions from the problem description
# What challenges did you face while handling missing data? How did you resolve them?
# - Faced challenges in deciding the imputation strategy (mean, median, mode, or more complex methods). Resolved by using SimpleImputer within the preprocessing pipeline, which is a standard and robust approach. For numerical features, mean imputation was chosen as a reasonable default. For categorical features, mode imputation is appropriate. Deciding on the best strategy often requires domain knowledge or further experimentation.
# Which machine learning algorithm performed the best, and why do you think it outperformed others?
# - Based on the R2 score on the training data, tree-based models like Random Forest or Gradient Boosting typically outperform Linear Regression on datasets with non-linear relationships or interactions between features. They can capture complex patterns more effectively. Gradient Boosting often performs slightly better than Random Forest as it builds trees sequentially, correcting errors of previous trees. However, the actual "best" model should ideally be determined on a separate validation set or through cross-validation to avoid overfitting.
# If you had access to additional data, what would you like to include, and how might it improve your model?
# - Additional data that would be valuable includes:
#   - Employee feedback/survey data on job satisfaction, work environment, manager support, and team dynamics.
#   - Project-specific data (e.g., project deadlines, complexity, team size).
#   - Work hours tracked per day/week.
#   - Commute time (if applicable for non-WFH).
#   - Historical performance data.
#   - Information about training or development opportunities.
#   - Data on employee personal situations (while respecting privacy and ethical considerations).
# - This data could provide a more holistic view of the factors contributing to burnout and potentially improve model accuracy by capturing more nuanced relationships. For instance, job satisfaction surveys directly measure employee sentiment, which is highly relevant to burnout. Project data could highlight high-pressure periods.
```
